### 3. Data Visualization

```python
# Visualization the average of Open based on time (Week)
import matplotlib.pyplot as plt
%matplotlib inline
plt.figure(figsize=(25, 25))

plt.subplot(3,3,1)
plt.ylabel('eth_etherprice')
df.eth_etherprice.plot()

plt.subplot(3,3,2)
plt.ylabel('eth_supply')
df.eth_supply.plot()

plt.subplot(3,3,3)
plt.ylabel('eth_marketcap')
df.eth_marketcap.plot()

plt.subplot(3,3,4)
plt.ylabel('eth_ens_register')
df.eth_ens_register.plot()

plt.subplot(3,3,5)
plt.ylabel('eth_hashrate')
df.eth_hashrate.plot()

plt.subplot(3,3,6)
plt.ylabel('eth_blocksize')
df['eth_blocksize'].plot()
```
